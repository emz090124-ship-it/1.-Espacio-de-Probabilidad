<div style="text-align: center;">
  <h1>Axiomas de Kolmog√≥rov</h1>
</div>

Sea $S$ el **espacio muestral** y $\mathbb{F}$ una **colecci√≥n de eventos** (subconjuntos de $S$).  
Una **funci√≥n de probabilidad** $P: \mathbb{F} \rightarrow [0,1]$ debe cumplir los siguientes axiomas fundamentales de la teor√≠a de la probabilidad:

### üîπ **1. Axioma de no negatividad**
Para todo evento $A \in \mathbb{F}$:
$$
P(A) \geq 0
$$
Esto significa que la probabilidad de cualquier evento nunca puede ser negativa.

### üîπ **2. Axioma de normalizaci√≥n**
La probabilidad del espacio muestral completo es igual a 1:
$$
P(S) = 1
$$
En otras palabras, la suma de las probabilidades de todos los resultados posibles es 1.

### üîπ **3. Axioma de aditividad finita**
Si $A_1, A_2, \dots, A_n \in \mathbb{F}$ son **eventos mutuamente excluyentes**
(es decir, $A_i \cap A_j = \emptyset$ para $i \neq j$), entonces:
$$
P\left( \bigcup_{i=1}^{n} A_i \right) = \sum_{i=1}^{n} P(A_i)
$$
Esto establece que la probabilidad de la uni√≥n de eventos disjuntos es la suma de sus probabilidades individuales.

> üí° **Importancia:**  
> Estos tres axiomas forman la base de la **teor√≠a cl√°sica de la probabilidad**, propuesta por Andrey Kolmog√≥rov en 1933, y contin√∫an siendo el pilar de la estad√≠stica moderna y el an√°lisis de incertidumbre.




```python
# ---------------------------------------------
# Verificaci√≥n de los Axiomas de Kolmog√≥rov
# ---------------------------------------------

# Espacio muestral
S = {'a', 'b', 'c'}

# Probabilidades de cada resultado
P = {'a': 0.2, 'b': 0.5, 'c': 0.3}

# --- Axioma 1: No negatividad ---
axioma1 = all(p >= 0 for p in P.values())

# --- Axioma 2: Normalizaci√≥n ---
axioma2 = abs(sum(P.values()) - 1) < 1e-8

# --- Axioma 3: Aditividad finita ---
A = {'a', 'b'}
B = {'c'}
union = A.union(B)
axioma3 = abs(sum(P[x] for x in union) - (sum(P[x] for x in A) + sum(P[x] for x in B))) < 1e-8

# Mostrar resultados
print("Verificaci√≥n de los Axiomas de Kolmog√≥rov\n")

print(f"1Ô∏è‚É£  Axioma de no negatividad: {'Cumple' if axioma1 else 'No cumple'}")
print(f"2Ô∏è‚É£  Axioma de normalizaci√≥n: {'Cumple' if axioma2 else 'No cumple'}")
print(f"3Ô∏è‚É£  Axioma de aditividad finita: {'Cumple' if axioma3 else 'No cumple'}")

print("\nEspacio muestral:", S)
print("Probabilidades:", P)
print("Suma total de probabilidades =", sum(P.values()))

```

$$
\text{\huge Secci√≥n I ‚Äî Enfoque Cl√°sico}
$$

---

El **enfoque cl√°sico de la probabilidad**, tambi√©n conocido como **probabilidad *a priori***, se basa en el supuesto de que todos los resultados posibles de un experimento son **igualmente probables**.  
Este modelo fue desarrollado formalmente en el siglo XVII por matem√°ticos como **Pierre-Simon Laplace**.

La probabilidad de un evento $A$ se define como:

$$
\mathbb{P}(A) = \frac{\text{N√∫mero de casos favorables}}{\text{N√∫mero total de casos posibles}}
$$

Este enfoque es √∫til en situaciones donde se puede enumerar claramente el conjunto de resultados posibles, como en **juegos de azar** (dados, monedas, cartas).  
Sin embargo, su aplicaci√≥n se limita cuando los resultados no son equiprobables o cuando no se puede determinar el espacio muestral con precisi√≥n.

> **Ejemplo:**  
> La probabilidad de obtener un n√∫mero par al lanzar un dado justo de seis caras es:

$$
\mathbb{P}(\text{par}) = \frac{3}{6} = 0.5
$$

###  **Espacio Muestral**

---

La **teor√≠a de la probabilidad** estudia los **fen√≥menos aleatorios**, es decir, aquellos experimentos cuyo resultado no puede preverse con certeza.  
Dado que no sabemos cu√°l ser√° el resultado de un experimento aleatorio, es fundamental **agrupar todos los posibles resultados** en un conjunto bien definido.

#### **Definici√≥n 1.1 ‚Äî Espacio Muestral**

El **espacio muestral**, tambi√©n llamado *espacio muestra*, es el conjunto de todos los posibles resultados de un experimento aleatorio.  
Se denota com√∫nmente por la letra griega $\Omega$ (omega may√∫scula).

Cada resultado individual del experimento se representa por $\omega$ (omega min√∫scula).

> **Ejemplo:**  
> Al lanzar un dado, el espacio muestral es:

$$
\Omega = \{1, 2, 3, 4, 5, 6\}
$$

Este concepto es esencial para construir modelos probabil√≠sticos dentro del **enfoque cl√°sico**, donde se asume que todos los resultados son igualmente probables.

### **Eventos**

---

Una vez definido el espacio muestral $(\Omega)$, podemos enfocarnos en los **eventos**, que son subconjuntos del espacio muestral. Un evento representa un conjunto de resultados que cumplen una cierta condici√≥n o caracter√≠stica del experimento aleatorio.

#### **Definici√≥n ‚Äî Evento**

Un **evento** es cualquier subconjunto del espacio muestral $(\Omega)$. Si el resultado del experimento pertenece al evento $(A)$, decimos que el evento **ocurre**.

> **Ejemplo:** Al lanzar un dado, el evento ‚Äúobtener un n√∫mero par‚Äù se representa como:
>
> $[
> A = \{2, 4, 6\}
> ]$

#### Tipos de eventos

- **Evento simple**: contiene un solo resultado.  
  Ejemplo: $\{3\}$

- **Evento compuesto**: contiene varios resultados.  
  Ejemplo: $\{1, 2, 3\}$

- **Evento imposible**: el conjunto vac√≠o $(\emptyset)$

- **Evento seguro**: el espacio muestral completo $(\Omega)$

Los eventos son la base para asignar probabilidades dentro del enfoque cl√°sico, donde se asume que todos los resultados son igualmente probables.

### **Probabilidad Condicional**

---

La **probabilidad condicional** es un concepto fundamental en el c√°lculo de probabilidades. Se utiliza para **ajustar la probabilidad de un evento** cuando se sabe que otro evento ha ocurrido. Este enfoque permite simplificar c√°lculos y entender mejor la relaci√≥n entre eventos.

#### Definici√≥n ‚Äî Probabilidad Condicional

Sean $(A)$ y $(B)$ dos eventos, y supongamos que $(\mathbb{P}(B) > 0)$. La **probabilidad condicional** del evento $(A)$, dado que ocurri√≥ el evento $(B)$, se denota por $(P(A \mid B))$ y se define como:

$$
\mathbb{P}(A \mid B) = \frac{\mathbb{P}(A \cap B)}{\mathbb{P}(B)}
$$

Esta f√≥rmula expresa la probabilidad de que ocurra $(A)$ **dentro del contexto** en que ya sabemos que ocurri√≥ $(B)$.

> **Ejemplo:**  
> Supongamos que en una clase, el $40\%$ de los estudiantes son mujeres y el $25\%$ de todos los estudiantes son mujeres que aprobaron el examen. Entonces, la probabilidad de que un estudiante haya aprobado **dado que es mujer** se calcula como:

$$
\mathbb{P}(\text{Aprob√≥} \mid \text{Mujer}) 
= \frac{\mathbb{P}(\text{Aprob√≥ y Mujer})}{\mathbb{P}(\text{Mujer})} 
= \frac{0.25}{0.40} 
= 0.625
$$

La probabilidad condicional es clave para construir modelos m√°s realistas y para aplicar el **Teorema de Bayes**, que se basa en esta definici√≥n.

---

### **Independencia de Eventos**

---

El concepto de **independencia** describe situaciones en las que la ocurrencia de un evento **no afecta** la probabilidad de ocurrencia de otro. Es una noci√≥n clave en la teor√≠a de la probabilidad, especialmente √∫til para **simplificar el c√°lculo de probabilidades conjuntas**.

#### **Definici√≥n ‚Äî Independencia**

Se dice que dos eventos $(A)$ y $(B)$ son **independientes** si se cumple la siguiente igualdad:

$$
\mathbb{P}(A \cap B) = \mathbb{P}(A) \cdot \mathbb{P}(B)
$$


Esta relaci√≥n implica que conocer que ocurri√≥ $(B)$ **no cambia** la probabilidad de que ocurra $(A)$, y viceversa.

> **Ejemplo:**  
> Al lanzar dos monedas justas, el evento $(A)$: ‚Äúla primera moneda cae en cara‚Äù y el evento $(B)$: ‚Äúla segunda moneda cae en cara‚Äù son independientes.

---

El enfoque geom√©trico es una **extensi√≥n del modelo cl√°sico** de probabilidad. En lugar de calcular la probabilidad de un evento mediante el conteo de casos favorables y posibles (cardinalidad), se utiliza una **medida geom√©trica** como el √°rea, el volumen o alguna otra caracter√≠stica espacial relevante, dependiendo del contexto del problema.

Este enfoque es especialmente √∫til cuando el espacio muestral no es discreto, sino continuo y representado dentro de un entorno geom√©trico como $(\mathbb{R}^2)$ o $(\mathbb{R}^3)$.

#### Definici√≥n ‚Äî Probabilidad Geom√©trica

Si un experimento aleatorio tiene como espacio muestral $(\Omega \subseteq \mathbb{R}^2)$, cuya **√°rea est√° bien definida y es finita**, entonces la **probabilidad geom√©trica** de un evento $(A \subseteq \Omega)$ se define como:

$$
\mathbb{P}(A) = \frac{\text{√Årea de } A}{\text{√Årea de } \Omega}
$$

Esta definici√≥n es v√°lida siempre que el concepto de √°rea del subconjunto \( A \) est√© bien definido.

---

> **Ejemplo: Problema del Encuentro**  
> Dos personas acuerdan encontrarse en un mismo lugar entre las $12:00$ y las $13:00$. Cada una llega en un instante aleatorio y permanece esperando $15$ minutos. ¬øCu√°l es la probabilidad de que se encuentren?

### Soluci√≥n 

El c√≥digo simula aleatoriamente los tiempos de llegada de dos personas durante una hora y calcula la proporci√≥n de veces que se encuentran, considerando que cada una espera 15 minutos. La funci√≥n `np.mean(np.abs(a-b) < 0.25)` estima la probabilidad de encuentro, mientras que el gr√°fico muestra visualmente los pares de llegadas y la l√≠nea donde las llegadas coinciden.

En Python:


```python
import numpy as np
import matplotlib.pyplot as plt

# Problema del encuentro
sim = 10_000
llegada_persona1 = np.random.rand(sim)
llegada_persona2 = np.random.rand(sim)

# Evaluar si se encuentran (diferencia menor a 15 minutos = 0.25 horas)
encontrarse = np.mean(np.abs(llegada_persona1 - llegada_persona2) < 0.25)
print(f"Probabilidad de encuentro: {encontrarse:.3f}")

# Visualizaci√≥n
plt.figure(figsize=(6,6))
plt.scatter(llegada_persona1 * 60, llegada_persona2 * 60, alpha=0.1, color='darkorange')
plt.plot([0, 60], [0, 60], color='navy', linestyle='--', label='x = y')
plt.title('Simulaci√≥n del problema del encuentro', fontsize=14, color='darkred')
plt.xlabel('Llegada Persona 1 (min)', fontsize=12)
plt.ylabel('Llegada Persona 2 (min)', fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', color='gray', alpha=0.5)
plt.tight_layout()
plt.show()
```

$$
\text{\huge Secci√≥n III ‚Äî Enfoque Frecuentista}
$$

---
El enfoque frecuentista define la probabilidad de un evento como el **l√≠mite de su frecuencia relativa** cuando el experimento aleatorio se repite muchas veces bajo condiciones id√©nticas.

Este modelo se basa en la observaci√≥n emp√≠rica: se registra cu√°ntas veces ocurre un evento $(A)$ en $(n)$ repeticiones del experimento, y se analiza c√≥mo evoluciona la proporci√≥n $(\frac{n_A}{n})$ a medida que $(n)$ crece.

#### Definici√≥n ‚Äî Probabilidad Frecuentista

Sean $(n_A)$ el n√∫mero de ocurrencias del evento $(A)$ en $(n)$ realizaciones de un experimento aleatorio. La **probabilidad frecuentista** de $(A)$ se define como:

$$
[
\mathbb{P}(A) = \lim_{n \to \infty} \frac{n_A}{n}
]
$$

---

Esta definici√≥n formaliza la idea de que la probabilidad refleja una **tendencia estable** en el largo plazo, y no una propiedad intr√≠nseca del evento.

> Ejemplo:  
> Si al lanzar una moneda 10,000 veces se obtiene cara en 4,980 ocasiones, entonces:

```python
import numpy as np

# N√∫mero de lanzamientos
lanzamientos = 10000

# Simulaci√≥n de lanzamientos: 0 = cruz, 1 = cara
resultados = np.random.choice([0, 1], size=lanzamientos, p=[0.5, 0.5])

# Contar cu√°ntas caras se obtuvieron
caras_obtenidas = np.sum(resultados)

# Probabilidad frecuentista
probabilidad_cara = caras_obtenidas / lanzamientos

print(f"Caras obtenidas: {caras_obtenidas}")
print(f"Probabilidad frecuentista de obtener cara: {probabilidad_cara:.4f}")
```
